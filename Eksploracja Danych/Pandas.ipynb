```python
import pandas as pd
import numpy as np
```

Reading DF

```python
df=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/cmc/cmc.data', names=['Age', 'Education', 'Husband_Education', 'Children', 'Religion', 'Working', 'Husband_working', 'Living_Index', 'Media_Exposure', 'Contraceptive_use'])
df
```

Normalize the Age attribute

```python
df.Age=(df.Age - df.Age.min()) / (df.Age.max() - df.Age.min())
df
```

Remove the column containing the information about religion

```python
df=df.drop('Religion',axis=1)
df
```

Create a second dataframe containig only women with highest degree of education

```python
df_highest=df[df.Education==4]
df_highest
```

Recode the Wife working now attribute, changing 0 to ‘no’ and 1 to ‘yes’

```python
df.Working=df.Working.astype('string')
def conver_to_str(ed):
    if ed=='1':
        return 'yes'
    else:
        return 'no'
df.Working=df.Working.apply(lambda x: conver_to_str(x))
df
```

Compute the frequency of contraceptive methods grouping by the woman’s education level

```python
df.groupby(df.Education).Contraceptive_use.value_counts()
```

Create a new column which will contain the average degree of education between a wife and a husband

```python
df['Avg_degree']=(df.Education+df.Husband_Education)/2
df
```

Create a new binary column which will indicate if a given woman has given birth to more than the average number of children

```python
average_kid=df.Children.mean()
def more_than_avg(number):
    if number > average_kid:
        return 1
    else:
        return 0
df['More_Kids']=df.Children.apply(lambda x : more_than_avg(x))
df
```

Change Standard of living column into an ordered category

```python
order=[1, 2, 3, 4]
df.Living_Index=df.Living_Index.astype(pd.CategoricalDtype(ordered=True))
df
```

Recode `Wife's education` attribute into a set of binary attributes using dummy variables

```python
education_dummies = pd.get_dummies(df['Education'], prefix='Education')

df = pd.concat([df, education_dummies], axis=1)
df=df.drop('Education',axis=1)
df
```

Plot the histogram of the number of children

```python
df.Children.plot(kind='hist')
```

Plot the bar chart of the standard of living

```python
df.Living_Index.value_counts().plot(kind='bar')
```
